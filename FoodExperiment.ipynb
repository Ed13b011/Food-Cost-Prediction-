# 1. TITLE

```python
#Training set
titles = list(df_train['TITLE'])
TITLE1 = []
TITLE2 = [] #maximum number titles in a cell, split into 2 columns
for i in titles:
    TITLE1.append(i.split(',')[0].strip().upper())
    try:
        TITLE2.append(i.split(',')[1].strip().upper())
    except:
        TITLE2.append('NONE')
        
#unique_titles.append('NONE') # appending NONE to Unique titles list


# test set
#Training set
titles2 = list(df_test['TITLE'])
TITLE3 = []
TITLE4 = [] #maximum number titles in a cell, split into 2 columns
for i in titles2:
    TITLE3.append(i.split(',')[0].strip().upper())
    try:
        TITLE4.append(i.split(',')[1].strip().upper())
    except:
        TITLE4.append('NONE')
```

# 3.RATING and VOTES

```python
#training set
ratings = list(df_train['RATING'])
for i in range(len(ratings)):
    try:
        ratings[i] = float(ratings[i])
    except:
        ratings[i] = np.nan

vote = list(df_train['VOTES'])
for i in range(len(vote)):
    try:
        vote[i] = int(vote[i].split(" ")[0].strip())
    except:
        pass
    
#test set
ratings2 = list(df_test['RATING'])
for i in range(len(ratings2)):
    try:
        ratings2[i] = float(ratings2[i])
    except:
        ratings2[i] = np.nan

vote2 = list(df_test['VOTES'])
for i in range(len(vote2)):
    try:
        vote2[i] = int(vote2[i].split(" ")[0].strip())
    except:
        pass

```

# 4. CITY and LOCALITY

```python
# training dataset 
#cities
cities1 = list(df_train['CITY'])
for i in range(len(cities1)):
    if type(cities1[i]) == float:
        cities1[i] = 'city'
        cities1[i] = cities1[i].strip().upper()
        
# locality
locality1 = list(df_train['LOCALITY'])
for i in range(len(locality1)):
    if type(locality1[i]) == float:
        locality1[i] = 'locality'
        locality1[i] = locality1[i].strip().upper()
        

# test dataset 
#cities
cities2 = list(df_test['CITY'])
for i in range(len(cities2)):
    if type(cities2[i]) == float:
        cities2[i] = 'city'
        cities2[i] = cities2[i].strip().upper()
        
# locality
locality2 = list(df_test['LOCALITY'])
for i in range(len(locality2)):
    if type(locality2[i]) == float:
        locality2[i] = 'locality'
        locality2[i] = locality2[i].strip().upper()
```

```python
# create new dataset for cleaned test data
cleaned_test = {}
cleaned_test['title3'] = TITLE3
cleaned_test['title4'] = TITLE4
cleaned_test['CUISINES9'] = cuisines9
cleaned_test['CUISINES10'] = cuisines10
cleaned_test['CUISINES11'] = cuisines11
cleaned_test['CUISINES12'] = cuisines12
cleaned_test['CUISINES13'] = cuisines13
cleaned_test['CUISINES14'] = cuisines14
cleaned_test['CUISINES15'] = cuisines15
cleaned_test['CUISINES16'] = cuisines16
cleaned_test['CITY'] = cities2
cleaned_test['LOCALITY'] = locality2
cleaned_test['RATING'] = ratings2
cleaned_test['VOTES'] = vote2
cleaned_test['RESTARANT_ID'] = df_test["RESTAURANT_ID"]
cleaned_test = pd.DataFrame(cleaned_test)
cleaned_test.head()
```

# Data Preprocessing

dealing with mising values, encoding categorical features, feature scaling


```python
# missing values 
# training set
cleaned_train.fillna(0, inplace = True)
# test set
cleaned_test.fillna(0, inplace = True)
#cleaned_train.isnull().sum(), cleaned_train.isnull().sum()
```

```python
# Dependent Variable
Y_train = cleaned_train.iloc[:,15].values  # 6 is the index of "Price" in the Training Set 

# Independent Variables
X_train = cleaned_train.iloc[:,cleaned_train.columns != 'COST'].values # selects all columns except "Price"

# Independent Variables for Test Set
X_test = cleaned_test.iloc[:,:].values
```

```python
# encoding categorical variables
from sklearn.preprocessing import LabelEncoder
#LE = LabelEncoder()
lE1 = LabelEncoder()
lE2 = LabelEncoder()

for i in range(14):
    X_train[:,i] = lE1.fit_transform(X_train[:,i])
    X_test[:,i] = lE2.fit_transform(X_test[:,i])

print(pd.DataFrame(X_train).head())
print(pd.DataFrame(X_test).head())
```

```python
#faeture scaling
from sklearn.preprocessing import StandardScaler
SC = StandardScaler()
X_train = SC.fit_transform(X_train)
X_test = SC.transform(X_test)

Y_train = Y_train.reshape(len(Y_train), 1)
Y_train = SC.fit_transform(Y_train)
Y_train = Y_train.ravel()
```

# Modelling

```python
from sklearn.svm import SVR

svr = SVR(kernel = "rbf")

svr.fit(X_train,Y_train)

Y_pred = SC.inverse_transform(svr.predict(X_test))

#pd.DataFrame(Y_pred, columns = ['COST']).to_excel("FoodSVR.xlsx", index = False)
```

```python
from sklearn.ensemble import RandomForestRegressor
model = RandomForestRegressor(n_jobs=-1)
model.fit(X_train,Y_train)
predRFR = SC.inverse_transform(model.predict(X_test))
pd.DataFrame(predRFR, columns = ['COST']).to_excel("FoodRFR2.xlsx", index = False)
print(pd.DataFrame(predRFR).head(10))
```
